```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt


weights = {}

# Loop through each column in the dataframe
for col in data.columns:
    # Check if the column has categorical data
    if data[col].dtype == 'object':
        # Calculate the weight of the column based on the value counts
        weight = data[col].value_counts().sum()
        # Add the column weight to the dictionary
        weights[col] = weight

# Create a WordCloud object with the column weights as input
wordcloud = WordCloud(width = 800, height = 800, background_color ='white', min_font_size = 10).generate_from_frequencies(weights)

# Display the WordCloud
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
  
plt.show()
```
